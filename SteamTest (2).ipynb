```python
import pandas as pd
df = pd.read_csv("steam.csv", encoding="latin-1")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].dtype=="object":
        df[col].fillna(df[col].mode()[0],inplace=True)
    elif (df[col].dtype in ["int64","float64"]) :
       df[col].fillna(0,inplace=True)
df.isnull().sum()

```

```python
df.duplicated()
```

```python
df=df.drop_duplicates()
df.duplicated()
```

```python
df.dtypes
```

```python
df["release_date"] = pd.to_datetime(df["release_date"], format="mixed", errors="coerce")
```

```python
df=df.rename(columns={"all_reviews":"reviews"})
df.columns
```

```python
df["minimum_requirements"] = df["minimum_requirements"].str.split(",", n=1).str[1] #n=1 pour agir sur le premier : seulement
df["recommended_requirements"] = df["recommended_requirements"].str.split(",", n=1).str[1] #n=1 pour agir sur le premier : seulement
df["reviews"] = df["reviews"].str.split(",", n=1).str[0] 
df.head()
```
